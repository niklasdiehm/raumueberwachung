```python
import pandas as pd
import matplotlib.pyplot as plt
temperature_data = pd.read_csv("temperature_data.csv")
```

```python
temperature_data.plot(x="timestamp", y="measurementValue")
```

```python
temperature_data["timestamp"] = pd.to_datetime(temperature_data["timestamp"])
```

```python
temperature_data.groupby([temperature_data.timestamp.dt.day, temperature_data.timestamp.dt.hour]).measurementValue.mean()
```
